## Introducao

O problema da mochila binaria consiste em escolher um subconjunto de itens de um conjunto de itens, de forma a maximizar o valor total dos itens escolhidos, respeitando a restricao de que a soma dos pesos dos itens escolhidos nao ultrapasse um valor maximo predefinido.

Na ultima aula exploramos eurisitcas aleatorias nessa aula exploraremos busca local.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
idealValue1 = 601
idealValue2 = 870
idealValue3= 648
idealValues = [idealValue1, idealValue2, idealValue3]
```

### Heuristica 1: Mochila Cheia
A primeira heuristica consiste na mesma heuristica aleatoria, mas garantindo que n√£o seja possivel adicionar mais itens a mochila.
Eu utilizei a heuristica de embaralhamento que ja conta com essa garantia.
Por se tratar da mesma heuristica, os resultados devem ser semelhantes.


### Resultados:


```python
randomValueSeries = []
for i in range(3):
	values = []
	with open("resultadosCheia"+str(i+1)+".txt") as f:
		for line in f:
			values.append(int(line.split()[1]))
		randomValueSeries.append(values)
```

```python
for i in range(3):
	plt.axhline(y = idealValues[i], color = 'r', linestyle = '--', label="Ideal"+str(i+1)) 
	plt.plot(randomValueSeries[i], label="Mochila Cheia"+str(i+1))
	plt.legend()
	plt.xlabel("Tentativa")
	plt.ylabel("Valor")
	plt.title("Cheia "+str(i+1))
	plt.show()
```

Como ja visto na aula anterior, a heuristica de embaralhamento nao e eficiente para resolver o problema da mochila binaria, pois a solucao encontrada e muito distante da solucao otima.

```python
for i in range(3):
	print("--------------------------------")
	arr = np.array(randomValueSeries[i])
	print("Cheia Entrada "+str(i+1))
	print("")
	print("Valor Ideal: ", idealValues[i])
	print("Media: ", np.mean(arr))
	print("erro percentual: ", (idealValues[i]- np.mean(arr) ) / idealValues[i] * 100)
	print("Melhor: ", np.max(arr))
	print("Pior: ", np.min(arr))

print("--------------------------------")
```

Podemos ver que o erro medio √© muito alto, e mesmo a melhor solucao encontrada pela heuristica aleatoria √© muito pior que a solucao otima.

### Heuristica 2: Substituicao
A segunda heuristica consiste em realizar a mochila cheia e depois substituir um item da mochila por um item que nao esta na mochila, caso isso melhore a solucao.
Ocorreu um bug que invalidou o resultado com a entrada 3 e infelizmente nao consegui corrigir a tempo.

### Heuristica 3: Hill Climb
A segunda heuristica consiste tambem em realizar substiuicoes. Testando alterar sempre a inclusao ou nao inclusao de um unico item por iteracao ate que nao haja mais melhora na solucao.

```python
hillValueSeries = []
for i in range(3):
	values = []
	with open("resultadosHill"+str(i+1)+".txt") as f:
		for line in f:
			values.append(int(line.split()[1]))
	hillValueSeries.append(values)
```

```python
for i in range(3):
	plt.axhline(y = idealValues[i], color = 'r', linestyle = '--', label="Ideal"+str(i+1)) 
	plt.plot(hillValueSeries[i], label="Hill Clibm"+str(i+1))
	plt.legend()
	plt.xlabel("Tentativa")
	plt.ylabel("Valor")
	plt.title("Hill"+str(i+1))
	plt.show()
```

```python
for i in range(3):
	arr = np.array(hillValueSeries[i])
	print("--------------------------------")
	print("hill Entrada "+str(i+1))
	print("")
	print("Valor Ideal: ", idealValues[i])
	print("Media: ", np.mean(arr))
	print("erro percentual: ", (idealValues[i]- np.mean(arr) ) / idealValues[i] * 100)
	print("Melhor: ", np.max(arr))
	print("Pior: ", np.min(arr))
print("--------------------------------")
```

Podemos ver que em algumas entradas o hill climbing n√£o consegue atingir um bom valor mas em outras simm isso provavelmente se deve a otimos locais que o algoritmo encontra.Para gerar a string binaria inicial eu gerei aleatoriamente porem cada bit tem 
90% de chance de ser 0. Isso evita a possibilidade de demorar muito para gerar inicialmente a solucao menor que W. Uma outra estrategia para gerar essa string inicial talvez funcionaria melhor.
